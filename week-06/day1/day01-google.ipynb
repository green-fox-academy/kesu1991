# What are the ten most recently updated apps?

```python
## Delete error row that the elements are not in the right column position
index_err = googleplaystore.index[googleplaystore['Last Updated'] == '1.0.19'].tolist()
index_err
updated_google_df = googleplaystore.drop(googleplaystore.index[[index_err[0]]])

# Define the column with date type
updated_google_df['Last Updated']= pd.to_datetime(updated_google_df['Last Updated'], format="%B %d, %Y") 

# Find the ten most recently updated apps
sorted = updated_google_df.sort_values(by=['Last Updated'], ascending=False)
sorted[:10][['App','Last Updated']]
```

# Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.

```python
# Check the top 25%, it is 4.5
sorted.describe(include = 'all')
sorted.head

# Create a column store the month and year
sorted['Month'] = sorted['Last Updated'].dt.month
sorted['Year'] = sorted['Last Updated'].dt.year
sorted[(sorted['Rating'] > 4.5) & (sorted['Month'] >= 7) & (sorted['Year'] == 2018)][['App', 'Rating', 'Last Updated']]
```

# Which app not to download? - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?

```python
# Delete the error value in the size and create a new dataset
sorted_update = sorted[sorted['Size'] != 'Varies with device']
sorted_update.head

# Extract number with size M
size_m = sorted_update['Size'].str.extract('(\d+|\d+\.\d)M')
size_m

# Create a list to store size M
size_m.columns = ['Size_M']
list_size_m = list(size_m.Size_M)
list_size_m[0:10]

# Add list to dataframe with name Size
sorted_update['Size_M'] = list_size_m
sorted_update

# Sort it based on Size
sorted_size = sorted_update.sort_values(by=['Size_M','Rating'], ascending=[False,True])[['App','Size','Rating']]
sorted_size
```

# Create a reviews reliability of reviews index, which takes into account, how many reviews does the app have and how high the rating is. What metric do you take? Create as a new field.

```python
# Sentiment_Polarity -1 to 1 poor to good
review_05 = review.index[review['Sentiment_Polarity'] == 1].tolist()
review.iloc[review_05]

# Join two tables 
df = pd.merge(sorted_update, review, on='App')

# Count how many reviews per app
df.groupby("App")["Translated_Review"].nunique().to_frame()

# Count average sentiment polarity
unsorted_sentiment = df.groupby("App")["Sentiment_Polarity"].mean().to_frame()
unsorted_sentiment

unsorted_rating = df.groupby("App")["Rating"].mean().to_frame()
unsorted_rating

list_rating = list(unsorted_rating.Rating)
unsorted_sentiment['Rating'] = list_rating

sorted_sentiment = unsorted_sentiment.sort_values(by=['Sentiment_Polarity', 'Rating'], ascending=[False,False])
sorted_sentiment
```
