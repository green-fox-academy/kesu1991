# Titanic ML Project

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline

# Loading test and train
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
test.info()
```

# Extract and clean the data set

```python
# Temperature (C), Humidity, Wind Speed (km/h)
knn_titanic_train = train[['Pclass','Sex','Age','Survived','Fare']]
knn_titanic_test = test[['Pclass','Sex','Age','Fare']]
knn_titanic_submit = test[['PassengerId','Pclass','Sex','Age','Fare']]

# Check NaN in 'train' set, Age contains null values!
knn_titanic_train[knn_titanic_train['Pclass'].isnull()]
knn_titanic_train[knn_titanic_train['Sex'].isnull()]
knn_titanic_train[knn_titanic_train['Age'].isnull()]
knn_titanic_train[knn_titanic_train['Survived'].isnull()]

# Check NaN in 'test' set, Age contains null values!
knn_titanic_test[knn_titanic_test['Age'].isnull()]

# Delete NaN from Age column in both 'train' and 'test' and 'submit'
knn_titanic_train = knn_titanic_train[knn_titanic_train['Age'].notnull()]
knn_titanic_train = knn_titanic_train[knn_titanic_train['Fare'].notnull()]
knn_titanic_train[knn_titanic_train['Age'].isnull()]
knn_titanic_train[knn_titanic_train['Fare'].isnull()]
knn_titanic_train.head()

knn_titanic_test = knn_titanic_test[knn_titanic_test['Age'].notnull()]
knn_titanic_test = knn_titanic_test[knn_titanic_test['Fare'].notnull()]

knn_titanic_submit = knn_titanic_submit[knn_titanic_submit['Age'].notnull()]
knn_titanic_submit = knn_titanic_submit[knn_titanic_submit['Fare'].notnull()]

# Create new column to indicate Sex with number 0 and 1 for both 'train' and 'test'
knn_titanic_train['Sex_N'] = np.where(knn_titanic_train['Sex']=='male', 0, 1)
knn_titanic_test['Sex_N'] = np.where(knn_titanic_test['Sex']=='male', 0, 1)
knn_titanic_train.head()
knn_titanic_test.head()
```

```python
# Defiavg(ne explanatory variable and our response
titanic_train_y = knn_titanic_train.iloc[:,3].values
titanic_train_x = knn_titanic_train.iloc[:,[0,2,4,5]].values
titanic_test_x = knn_titanic_test.iloc[:,[0,2,3,4]].values
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(titanic_train_x)

titanic_train_x = scaler.transform(titanic_train_x)  
titanic_test_x = scaler.transform(titanic_test_x)
```

# Training and modelling

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(titanic_train_x, titanic_train_y)  
```

# Prediction

```python
Survived = classifier.predict(titanic_test_x)
```

# Output the result

```python
from pandas import DataFrame

# Store PassengerID into array
PassengerId = knn_titanic_submit.iloc[:,0].values
list(PassengerId)

# Create df
export_df = {'PassengerId': list(PassengerId), 'Survived': list(Survived)}

df = DataFrame(export_df, columns= ['PassengerId', 'Survived'])
```

```python
df.head()
```

```python
df.to_csv (r'1.csv', index = None, header=True)
```

```python
# Temperature (C), Humidity, Wind Speed (km/h)
knn_titanic_noage_train = train[['Pclass','Sex','Age','Survived']]
knn_titanic_noage_test = test[['Pclass','Sex','Age']]
knn_titanic_noage_submit = test[['PassengerId','Pclass','Sex','Age']]

# Create a data set that without age to test knn only with sex and class
knn_titanic_noage_train = knn_titanic_noage_train[knn_titanic_noage_train['Age'].isnull()]
knn_titanic_noage_test = knn_titanic_noage_test[knn_titanic_noage_submit['Age'].isnull()]
knn_titanic_noage_submit = knn_titanic_noage_submit[knn_titanic_noage_submit['Age'].isnull()]

# Create new column to indicate Sex with number 0 and 1 for only sex and class knn model
knn_titanic_noage_train['Sex_N'] = np.where(knn_titanic_noage_train['Sex']=='male', 0, 1)
knn_titanic_noage_test['Sex_N'] = np.where(knn_titanic_noage_test['Sex']=='male', 0, 1)

# Define explanatory variable and our response
titanic_noage_train_y = knn_titanic_noage_train.iloc[:,3].values
titanic_noage_train_x = knn_titanic_noage_train.iloc[:,[0,4]].values
titanic_noage_test_x = knn_titanic_noage_test.iloc[:,[0,3]].values

# Scaling
scaler1 = StandardScaler()  
scaler1.fit(titanic_noage_train_x)

titanic_noage_train_x = scaler1.transform(titanic_noage_train_x)  
titanic_noage_test_x = scaler1.transform(titanic_noage_test_x)
```

```python
# Prediction
classifier = KNeighborsClassifier(n_neighbors=5)  
classifier.fit(titanic_noage_train_x, titanic_noage_train_y)
```

```python
# Predict
Survived = classifier.predict(titanic_noage_test_x)

# Store PassengerID into array
PassengerId = knn_titanic_noage_submit.iloc[:,0].values

# Create df
export_noage_df = {'PassengerId': list(PassengerId), 'Survived': list(Survived)}

df1 = DataFrame(export_noage_df, columns= ['PassengerId', 'Survived'])
```
